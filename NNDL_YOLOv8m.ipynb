```python
import pandas as pd

df1 = pd.read_csv("/content/runs/detect/train5/results.csv")
df2 = pd.read_csv("/content/runs/detect/train6/results.csv")
df3 = pd.read_csv("/content/runs/detect/train7/results.csv")
df4 = pd.read_csv("/content/runs/detect/train8/results.csv")
df5 = pd.read_csv("/content/runs/detect/train10/results.csv")
```
